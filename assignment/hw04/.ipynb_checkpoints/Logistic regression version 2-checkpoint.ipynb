```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
```

```python
# the logistic function
def logistic_func(theta, x):    
    t = x.T @ theta    
    g = np.zeros(t.shape)    
    # split into positive and negative to improve stability    
    g[t>=0.0] = 1.0 / (1.0 + np.exp(-t[t>=0.0]))     
    g[t<0.0] = np.exp(t[t<0.0]) / (np.exp(t[t<0.0])+1.0)    
    return g
# function to compute the gradient of -f
def grad(theta, x, y):    
    g = logistic_func(theta,x)    
    return -(x @ (y-g))

# function to compute output of LR classifier
def lr_predict(theta,x):    
    # form Xtilde for prediction    
    x = np.vstack((x.T , np.ones(x.shape[0])) )    
    return logistic_func(theta,x)

#the objective function -f
def f_eval(theta, x, y):    
    t = x.T @ theta    
    return -np.vdot(t,y) + np.sum(np.log(1+np.exp(t)))

#function to evaluate grad alphadef 
def grad_alpha(theta, x, y , alpha, d): 
    return grad(theta + alpha*d,x,y) @ d.T
```

```python
# gradient descent# returns theta and number of iterations
def gradDesc(x,y,alpha,delta,maxiter):    
    # Initialization    
    theta = np.zeros(x.shape[0])    
    d = -grad(theta, x, y)    
    k = 0    
    while (k < maxiter) and (np.linalg.norm(d) > delta):        
        theta = theta + alpha*d        
        d = -grad(theta, x, y)        
        k = k+1    
    return theta, k
```

# standard backtracking algorithm

```python
def backtracking(theta,x,y):
    alpha=1
    rho=0.9
    c1=0.5
    max1=1000
    n=0
    d=-grad(theta, x, y)
    while(f_eval(theta+alpha*d,x,y)>f_eval(theta,x,y)+c1*alpha*np.vdot(d,-d) and n<max1):
        alpha*=rho
        n+=1
    return alpha,n

def gradDesc_backtrack(x,y,alpha,delta,maxiter):
     # Initialization    
    theta = np.zeros(x.shape[0])    
    d = -grad(theta, x, y)    
    k = 0    
    while (k < maxiter) and (np.linalg.norm(d) > delta):
        alpha,n=backtracking(theta,x,y)
        theta = theta + alpha*d        
        d = -grad(theta, x, y)        
        k = k+1    
    return theta, k
```

# the heavy ball algorithm

```python
def graDesc_heavyball(x,y,delta,maxiter):
    alpha=1
    beta=0.92
    i=0
    iteration=0
    theta=np.zeros(x.shape[0])
    theta0=theta
    d= -grad(theta, x, y)
    while((i<maxiter) and np.linalg.norm(d)>delta):
        theta1=theta
        alpha,n=backtracking(theta,x,y)
        theta=theta+beta*(theta-theta0)+alpha*d  #the value of Xk
        theta0=theta1 #the value of Xk-1
        i+=1
        iteration+=n
        d=-grad(theta, x, y)
    return theta, i
```

# Nesterov's method

```python
def graDesc_Nesterov(x,y,alpha,delta,maxiter):
    k=1 #initialize k
    beta=(k-1)/(k+2)
    theta=np.zeros(x.shape[0])
    theta0=theta
    iteration=0
    i=0
    d=-grad(theta,x,y)
    while((i<maxiter) and (np.linalg.norm(d)>delta)):
        theta1=theta
        p=beta*(theta-theta0)
        theta=theta+p-alpha*grad(theta+p,x,y)  #the value of Xk
        theta0=theta1 #the value of Xk-1
        iteration+=1
        i+=1
        k+=1
        beta=(k-1)/(k+2)
        d=-grad(theta,x,y)
    return theta, iteration
```

# Newton's method

```python
def sigmoid(x):
    return 1.0/(1+np.exp(-x)) 

def hessian(th,x,y):
    xTrans = x.transpose()                                      
    sig = sigmoid(np.dot(xTrans,th))                             
    result = (1.0/len(xTrans) * x.dot(np.diag(1 - sig)* np.diag(sig)).dot(x.T))  
    return result

def graDesc_Newton(x,y,alpha,delta,maxiter):
     # Initialization    
    theta = np.zeros(x.shape[0])
    h=np.linalg.inv(hessian(theta,x,y))
    d = -h@(grad(theta, x, y))    
    k = 0    
    while (k < maxiter) and (np.linalg.norm(d) > delta):
        
        theta = theta + alpha*d    
        h=np.linalg.inv(hessian(theta,x,y))
        d = -h@(grad(theta, x, y))
        k = k+1    
    return theta, k

def graDesc_Newton_backtracking(x,y,delta,maxiter):
     # Initialization    
    theta = np.zeros(x.shape[0])
    alpha=1.0
    h=np.linalg.inv(hessian(theta,x,y))
    g=grad(theta, x, y)
    d = -h @ g  
    k = 0    
    while (k < maxiter) and (np.linalg.norm(d) > delta):
        alpha,n=backtracking(theta,x,y)
        theta = theta + alpha*d 
        h=np.linalg.inv(hessian(theta,x,y))
        d = -h @ (grad(theta, x, y))
        k = k+1    
    return theta, k
```

# BFGS method

```python
def gradDesc_BFGS(x,y,delta,maxiter):
         # Initialization    
    theta = np.zeros(x.shape[0])
    alpha=1.0
    h=np.linalg.inv(hessian(theta,x,y))
    g=grad(theta, x, y)
    d = -h @ g  
    k = 0    
    while (k < maxiter) and (np.linalg.norm(d) > delta):
        theta0=theta
        g0=g
        alpha,n=backtracking(theta,x,y)
        theta = theta + alpha*d
        g=grad(theta, x, y)
        s=np.mat(theta-theta0).T
        r=np.mat(g-g0).T
        a=h @ r
        gama=s.T @ r
        print(gama)
        h=h+np.array((gama+(r.T @ a))/gama**2)*np.array((s @ s.T))-np.array((a @ s.T)/gama)-np.array((s @ a.T)/gama)
        d = -h @ g
        k = k+1    
    return theta, k   
```
